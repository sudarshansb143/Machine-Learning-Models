```python
import numpy as np
import pandas as pd
import warnings
warnings.defaultaction = "ignore"
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
c_data = load_breast_cancer()
```

```python
df = pd.DataFrame(c_data.data, columns=c_data.feature_names)
```

```python
df["target"] = c_data.target
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn")
```

```python
sns.scatterplot(df["mean radius"], df["mean texture"], )
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 9)
plt.rcParams['figure.dpi'] = 120
sns.set()
plt.xlabel("mean Radius")
plt.ylabel("mean texture)")
sns.scatterplot(x = df["mean radius"], y = df["mean texture"], hue=df.target.map({0: "malignant", 1: "benign"}), data=df)
plt.show()
```
