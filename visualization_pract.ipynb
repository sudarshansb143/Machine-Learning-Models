```
temp = [i.strip().lower().replace(" ", "_") for i in df.region]
df.region = temp
```

```
tips = sns.load_dataset("tips")
tips.head()
```

```
sns.scatterplot(x="total_bill", y= "tip", data = tips, hue = "smoker")
```

```
# Setting the hue order
sns.scatterplot(x="total_bill", y= "tip", data = tips, hue = "smoker", hue_order=["Yes", "No"])
```

```
# changing the color pallete
hue_color = {
    "Yes" : "red",
    "No" : "green"
}
sns.scatterplot(x="total_bill", y= "tip", data = tips, hue = "smoker", palette=hue_color)
plt.show()
```

```
# Seeting hue to count plot
# Plotting the smokers  in which those are male and female
sns.countplot(x="smoker", data = tips, hue = "sex")
plt.show()

sns.countplot(y="smoker", data = tips, hue = "sex")
plt.show()
```

```
# creating the subplot 
sns.relplot(x = "total_bill", y = "tip", data = tips ,  kind = "scatter", col = "smoker")
plt.show()
# Conclusion is the non smoker persons  tends to give the more tip on the  total bill than that of the smoker ones  
```

```
sns.relplot(x = "total_bill", y = "tip", data = tips ,  kind = "scatter", row = "smoker")
plt.show()
```

```
sns.relplot(x = "total_bill", y = "tip", data = tips ,  kind = "scatter", col = "smoker", row = "sex")
# Conclusion further is even though the non smoker gives more tip but in that also the men gives the more tip than that of the females
```

```
# custom order of the both smoker and sex
sns.relplot(x = "total_bill", 
            y = "tip", data = tips,  
            kind = "scatter", 
            col = "smoker",
            col_order = ["Yes", "No"], 
            row = "sex",
            row_order = ["Male", "Female"])
```

```
# Adding the customizations to plot
sns.relplot(x="total_bill", y = "tip", data = tips, size = "size", kind = "scatter", hue = "size")
```
